# Contextual Embeddings (BERT Example)

## Contextual embeddings ka matlab hota hai:

## ek word ka meaning depend karta hai sentence ke context par.

## Example:

## “bank” in “I went to the bank to deposit money” → financial bank 🏦

## “bank” in “I sat by the river bank” → nadi ka kinara 🌊

## Dono jagah “bank” same word hai, lekin context alag hai, isliye BERT dono ka alag vector (embedding) deta hai.
## Yeh hi contextual embedding hoti hai.

```python
from transformers import BertTokenizer, BertModel
import torch

# Load pretrained model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Input
text = "I love Pakistan"
inputs = tokenizer(text, return_tensors='pt')

# Get embeddings
outputs = model(**inputs)
embeddings = outputs.last_hidden_state

print(embeddings.shape)  # [1, sequence_length, 768]
```

# word 2 vec embedding

## Word2Vec ek word embedding algorithm hai jo words ko numbers (vectors) mai convert karta hai.
##  Matlab har word ka ek numeric representation (embedding) banta hai jo uske meaning ke paas-paas wale words ke saath milta-julta hota hai.

## Example:

## “king” aur “queen” ke vectors paas honge

## “car” aur “bus” ke vectors bhi paas honge

## “king − man + woman ≈ queen” → yeh relation bhi Word2Vec pakad sakta hai

```python
from gensim.models import Word2Vec
sentences = [
    ["i", "love", "machine", "learning"],
    ["word2vec", "creates", "word", "embeddings"],
    ["deep", "learning", "is", "powerful"],
    ["i", "love", "natural", "language", "processing"]
]
skip = Word2Vec(sentences,vector_size=10,window=3,min_count=1)
print("skip embedding")
print(skip.wv["learning"])

# cbow (continoues bag of words)
cbow = Word2Vec(sentences,vector_size=10,window=3,min_count=1)
print("cbow vector for learning")
print(cbow.wv["learning"])

# Similarity Words
print("\n Most similar worrds to 'love':")
print(skip.wv.most_similar("love"))
```

## Type of word 2 vec	
## CBOW	Continuous Bag of Words	Context (around words) se target word predict karta hai
## Skip-Gram	Skip-Gram	Target word se uske aas-paas ke words (context) predict karta hai

# 2. CBOW (Continuous Bag of Words)

```python
from gensim.models import Word2Vec

# Corpus
sentences = [
    ["i", "love", "learning", "ai"],
    ["deep", "learning", "is", "fun"],
    ["i", "enjoy", "machine", "learning"]
]

# CBOW Model (sg=0 means CBOW)
cbow_model = Word2Vec(
    sentences,
    vector_size=50,
    window=2,
    min_count=1,
    sg=0  # CBOW
)

# Embedding for 'learning'
print(cbow_model.wv['learning'])

# Similar words
print(cbow_model.wv.most_similar('learning'))
```

# example 

```python
Sentence: "I love machine learning"

👉 CBOW:

Model seekhta hai —
context: (“I”, “machine”) → target: “love”
```

# 🔹 3. Skip-Gram Model

```python
from gensim.models import Word2Vec

sentences = [
    ["i", "love", "learning", "ai"],
    ["deep", "learning", "is", "fun"],
    ["i", "enjoy", "machine", "learning"]
]

# Skip-Gram Model (sg=1 means skip-gram)
skipgram_model = Word2Vec(
    sentences,
    vector_size=50,
    window=2,
    min_count=1,
    sg=1  # Skip-Gram
)

# Embedding for 'learning'
print(skipgram_model.wv['learning'])

# Similar words
print(skipgram_model.wv.most_similar('learning'))
```

# example 

```python
Sentence: "I love machine learning"

👉 Skip-Gram:

Model seekhta hai —
target: “love” → context: (“I”, “machine”)
```

# 4. Visualizing Word Embeddings (optional cool step 😎)

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Use trained model (e.g., skipgram_model)
words = list(skipgram_model.wv.key_to_index.keys())
vectors = [skipgram_model.wv[w] for w in words]

pca = PCA(n_components=2)
result = pca.fit_transform(vectors)

plt.figure(figsize=(6,6))
plt.scatter(result[:,0], result[:,1])

for i, word in enumerate(words):
    plt.annotate(word, xy=(result[i,0], result[i,1]))
plt.show()
```

# Glove embedding

### GloVe (Global Vectors for Word Representation) ek word embedding model hai jo Stanford University ne banaya tha.
### yeh bhi Word2Vec ki tarah words ko numbers (vectors) mai convert karta hai — lekin training ka concept alag hai.

## Word2Vec context ke through embeddings seekhta hai (CBOW/SkipGram).
## Lekin GloVe sentence-based nahi balki global statistics (word co-occurrence matrix) par based hota hai.

## Simple Samajh:

## GloVe ek matrix banata hai jisme har cell batata hai ki ek word kitni baar doosre word ke paas aata hai.

## Fir mathematical optimization se har word ka vector (numbers ka set) seekh leta hai jisse yeh matrix reconstruct ho sake.

## Matlab:

## Agar “king” aur “queen” aksar “royal”, “crown”, “palace” ke aas-paas aate hain,
## to inke embeddings similar honge.

```python
import gensim.downloader as api

# Pretrained GloVe model download (50-dim)
glove_model = api.load("glove-wiki-gigaword-50")  

# Vector for a word
vector = glove_model["king"]
print("Embedding for 'king':", vector)

# Similar words
similar = glove_model.most_similar("king", topn=5)
print("\nMost similar words to 'king':")
for word, score in similar:
    print(word, score)

# Cosine similarity between two words
sim = glove_model.similarity("king", "queen")
print("\nSimilarity between 'king' and 'queen':", sim)
```

# Sentence Embedding (Sentence-BERT Example)

## Sentence embedding ka matlab hai — poore sentence ka meaning ek single vector (number array) mai convert karna.

## Matlab:
## instead of har word ka embedding (jaise BERT karta hai),
## SentenceTransformer poore sentence ka ek combined semantic vector deta hai.

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = ["I love Pakistan", "Pakistan is beautiful"]
embeddings = model.encode(sentences)

print(embeddings.shape)  # (2, 384)
```
